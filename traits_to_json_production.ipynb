```python
"""
Inputs: 
    trait sheet with traits for each NFT
    IPFS prefix for male and female images
    image file suffix

This can easily be modified to shift or shuffle the metadata using an integer seed

"""
import pandas as pd
import json
#import random #do we want to seed this?
#import numpy as np
import re
import os
#SET CONSTANTS
TRAIT_SHEET='shuffled_vigilante_testD_sheet.csv'
MALE_IMAGE_IPFS_PREFIX = 'ipfs://QmQRjaue8SwCw45y6Txyc4KAM5A2BdqyakGj6rNKeoNgVT/'
FEMALE_IMAGE_IPFS_PREFIX = 'ipfs://QmQRjaue8SwCw45y6Txyc4KAM5A2BdqyakGj6rNKeoNgVT/'
FEMALE_SUFFIX = '/base_f.png'
MALE_SUFFIX = '/base.png'

DESCRIPTION = '3,777 Vigilantes on a mission to make NFT launches fair and equitable. Join us. https://vigilante.eth.limo/'

curpath = os.path.abspath(os.curdir)
FILE_PATH = curpath+'/metadata'

#load traits

trait_sheet=pd.read_csv(TRAIT_SHEET)


def stdTrait(str):
    str = re.sub(r"(_|-)", " ", str)
    str = str.title()
    return str
      
# function to convert strings to json format and saves the file as .json
def save_metadata(nft_metadata):
    attributes = []
    #note 'Backgrounds' is being changed to 'Background' here
    male_trait_labels = {'Mask': 'm_mask', 'Headgear': 'm_headgear','Outfit':'m_outfit','Accessory':'m_accessories','Eyes':'Eyes','Background':'Backgrounds','Gender':'Gender'}
    female_trait_labels = {'Mask': 'f_mask', 'Headgear': 'f_headgear','Outfit':'f_outfit','Accessory':'f_accessories','Eyes':'Eyes','Background':'Backgrounds','Gender':'Gender'}

    for attribute in ['Mask', 'Outfit', 'Eyes','Headgear','Accessory','Gender','Background']:
            #if it's a male NFT give male traits
            if nft_metadata['Gender']=='male':
                attributes.append({
                    "trait_type": stdTrait(attribute),
                    "value": stdTrait(nft_metadata[male_trait_labels[attribute]])
                })
            else:#if it's a female NFT give female traits
                 attributes.append({
                    "trait_type": stdTrait(attribute),
                    "value": stdTrait(nft_metadata[female_trait_labels[attribute]])
                })               

    metadata = {
        "name": 'Vigilante #' + str(nft_metadata['TOKEN ID']),
        "description": DESCRIPTION,
        "attributes": attributes}
    
    if nft_metadata['Gender']=='male':
        metadata['image']=MALE_IMAGE_IPFS_PREFIX+str(nft_metadata['Number'])+MALE_SUFFIX
    else:
        metadata['image']=FEMALE_IMAGE_IPFS_PREFIX+str(nft_metadata['Number'])+FEMALE_SUFFIX
        
    output_file=str(nft_metadata['TOKEN ID'])
    complete_name = os.path.join(FILE_PATH, output_file)
    with open(complete_name, 'w') as outfile:
        json.dump(metadata, outfile, indent=4)

        
        
            
        
```

```python

for nft in range(0,len(trait_sheet)):
    save_metadata(trait_sheet.loc[nft])
```
