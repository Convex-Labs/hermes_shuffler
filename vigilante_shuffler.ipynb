```python
"""
Inputs NFT metadata, shuffles it, outputs shuffled metadata

Inputs: 
trait_sheet: a csv with columns:
Number,Eyes,Backgrounds,m_mask,m_headgear,m_outfit,m_accessories,Token ID,f_mask,f_headgear,f_outfit,f_accessories,gender

seed: a 256 bit integer generated by the chainlink VRF

"""

TRAIT_SHEET_NAME='vigilante_testD_sheet.csv'

#chainlink gives us a 256-bit int but pandas can only handle an int up to 2^32-1 so we take modulo here
SEED=41288263153917951879409839609720033986319181727292631477037293209191196903743%(2**32-1)

import pandas as pd
#load traits
trait_sheet_unshuffled=pd.read_csv(TRAIT_SHEET_NAME)

#shuffle
trait_sheet_shuffled=trait_sheet_unshuffled.sample(frac=1,random_state=SEED).reset_index(drop=True)
#add token ID to sheet
trait_sheet_shuffled['TOKEN ID']=trait_sheet_shuffled.index+1
#output shuffled sheet
trait_sheet_shuffled.to_csv('shuffled_'+TRAIT_SHEET_NAME)
```
