```python
"""
Inputs NFT metadata, shuffles it, outputs shuffled metadata

Inputs: 
trait_sheet: a csv with columns:
Image_Number,Eyes,Backgrounds,Mask,Headgear,Outfit,Accessory_M,Accessory_F,Gender

seed: a 256 bit integer generated by the chainlink VRF
Obtain seed by calling s_randomWords(0) on 0xbD1d2Ea3127587f4ECFD271e1dADFc95320b8DeA
for the first reveal. 

"""
#copy and paste in the filename of the metadata file downloaded from IPFS with no file ending
TRAIT_SHEET_NAME='committed_HonestNFT_Vigilante_Metadata.csv'

#chainlink gives us a 256-bit int but pandas can only handle an int up to 2^32-1 so we take modulo here
#seed beginning with 987188 is an example. 
SEED=98718832011749334867803141546608476580547334638907568821983820441926603690513%(2**32-1)

import pandas as pd
#load traits, skip first line because it contains hash of this file
trait_sheet_unshuffled=pd.read_csv(TRAIT_SHEET_NAME,skiprows=1)

#shuffle
trait_sheet_shuffled=trait_sheet_unshuffled.sample(frac=1,random_state=SEED).reset_index(drop=True)
#add token ID to sheet
trait_sheet_shuffled['TOKEN ID']=trait_sheet_shuffled.index+1
#output shuffled sheet
trait_sheet_shuffled.to_csv('shuffled_'+TRAIT_SHEET_NAME)
```
